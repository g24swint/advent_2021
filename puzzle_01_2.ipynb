```python
df = pd.read_csv("puzzle_01_1_input.txt", header=None)
df.columns=["depth"]
with pd.option_context("display.max_rows", 25, "display.min_rows", None):
    display(df)
```

```python
with pd.option_context("display.max_rows", 25, "display.min_rows", None):
    df["rolling_sum"] = df.depth.rolling(3, min_periods=3, ).sum()
    df["delta"] = df.rolling_sum.diff()
    df["increase_flag"] = df.delta > 0
    display(df)
```
